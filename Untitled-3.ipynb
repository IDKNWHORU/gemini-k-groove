```python
import json
from transformers import BertTokenizer, BertModel
import numpy as np
import torch # 추가된 코드

# JSON 파일 읽기
with open("expert_poses.json", "r") as f:
    data = json.load(f)

# BERT 토크나이저 및 모델 로딩
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

# 임베딩 벡터 생성
embeddings = []
for item in data:
    text = json.dumps(item)  # JSON 데이터를 텍스트로 변환
    encoded_text = tokenizer.encode_plus(
        text,
        add_special_tokens=True,
        max_length=512,
        truncation=True,
        padding="max_length",
        return_tensors="pt",
    )
    with torch.no_grad():
        output = model(**encoded_text)
        embeddings.append(output.last_hidden_state[:, 0, :].numpy())  # 첫 번째 토큰의 임베딩 벡터 사용

# 임베딩 벡터를 .npy 파일로 저장
embeddings = np.array(embeddings)

print(embeddings)
np.save("expert_pose_embeddings.npy", embeddings)
```
